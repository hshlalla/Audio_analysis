```python
def paths_and_labels_to_dataset(audio_paths):
    """Constructs a dataset of audios and labels."""
    path_ds = tf.data.Dataset.from_tensor_slices(audio_paths)
    audio_ds = path_ds.map(lambda x: path_to_audio(x))
    return tf.data.Dataset(audio_ds)


def path_to_audio(path):
    """Reads and decodes an audio file."""
    audio = tf.io.read_file(path)
    audio, _ = tf.audio.decode_wav(audio, 1, 16000)
    print(audio)
    y=np.concatenate((audio,np.zeros(10000)),axis=0)
    slices = int(y.shape[0] / orig_sr)
    sample = tf.split(y[: slices * orig_sr], slices)
    audio=tf.stack(sample)
    return audio
```
